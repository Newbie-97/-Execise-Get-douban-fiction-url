```python
import re
import requests

s = requests.Session()

def getHTML(url , code='utf-8'):
    try:
        response = s.get(url)
        response.raise_for_status()
        response.encoding = code
        return response.text
        print(response.status_code)
    except:
        print('getHTML Error')

url = 'https://read.douban.com/kind/100?start=0&sort=hot&promotion_only=False&min_price=None&max_price=None&works_type=None'
html = getHTML(url)
print(html)
```
